### calling the inplace parameter is for changing the actual table/pd. for example:

```python
table = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
table

# drop without inplace- changes but not the original
new_table = table.drop('A', axis=1)
table

# with inpace- changes the origin
new_table_2 = table.drop('A', axis=1, inplace=True)
table
```

## אינדקס הוא דמוי מפתח
כלומר: יכולים להיות כמה אינדקסים דומים- אך עמודת אינדקס יש רק אחת והיא אמצעי הזיהוי של כל שורה בעמודות האחרות. זה בעצם כמו העמודה הראשית.
זיהוי העמודה הראשית היא על פי הכותרת- ולכן היא לא יכולה להיות שווה בשמה לכותרות של עמודות אחרות

axis0- rows
axis1- columns

```python
# & == and
df[(df['International plan'] == 'Yes') & (df['Churn'] == True)].index
# | == or
df[(df['International plan'] == 'Yes') | (df['Churn'] == True)].index
# ~ == not
df[~(df['International plan'] == 'Yes')].index
# isin
df[df['State'].isin(['LA', 'WY'])]
# between
df[df['Account length'].between(50, 100)]
```
